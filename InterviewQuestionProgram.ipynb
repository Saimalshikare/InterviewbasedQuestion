```python
# Print all prime number to given range
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
n = int(input("Enter a number: "))
for num in range(2, n + 1):
    if is_prime(num):
        print(num)
```

```python
# write a function to find the longest word in senetence
def longest_find(sentence):
    result=sentence.split()
    if not result:
        return''
    return max(result,key=len)
longest_find('sai ravindra malshikare')
```
